```python
import pandas as pd
```

```python
medical = pd.read_csv("insurance.csv")
medical
```

```python
import plotly.express as px
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
fig = px.histogram(medical, x="age", nbins=30, title="Distribution of Age")
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(medical, x="bmi", nbins=30, title="Distribution of Age")
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(medical, x="age", color="charges", title="Distribution of Age")
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(medical, x="charges", color="smoker", title="Distribution of Age")
fig.update_layout(bargap=0.1)
fig.show()
```
